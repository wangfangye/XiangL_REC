```python
from collections import Counter
def counter(arr):
    return Counter(arr).most_common() # 返回出现频率最高的两个数
# 结果：[(2, 3), (1, 2)]
```

```python
counter(allCrime_np[:,4]==152)
```

```python
ac_152 = allCrime_np[allCrime_np[:,4]==152,5]
```

```python
def counter2(label):
    ac = allCrime_np[allCrime_np[:,4]==label,5]
    print(len(ac))
    return Counter(ac).most_common()
```

```python
gb_cri_np = gb_cri.values
```

```python
cri2law = []
for label,count in gb_cri_np:
    ac2 = counter2(label)
    cri2law.append([label,CRIME.accu[label],count,ac2])
```
